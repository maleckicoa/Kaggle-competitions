```python
# Restore or create MLflow experiment named 'demo_experiment' and set it active
import mlflow
from mlflow.tracking import MlflowClient
from mlflow.entities import ViewType

TARGET_EXPERIMENT_NAME = "demo_experiment"

client = MlflowClient()  # Uses current tracking URI
print(f"Tracking URI: {mlflow.get_tracking_uri()}")

# Find experiment by name, including deleted ones
matches = [
    exp for exp in client.search_experiments(view_type=ViewType.ALL)
    if exp.name == TARGET_EXPERIMENT_NAME
]

if matches:
    exp = matches[0]
    if getattr(exp, "lifecycle_stage", None) == "deleted":
        client.restore_experiment(exp.experiment_id)
        print(f"Restored experiment '{TARGET_EXPERIMENT_NAME}' (id={exp.experiment_id}).")
    else:
        print(f"Experiment '{TARGET_EXPERIMENT_NAME}' exists (id={exp.experiment_id}).")
else:
    print(f"No experiment named '{TARGET_EXPERIMENT_NAME}' found; it will be created.")

mlflow.set_experiment(TARGET_EXPERIMENT_NAME)
active = client.get_experiment_by_name(TARGET_EXPERIMENT_NAME)
print(
    f"Active experiment set: name='{active.name}', id='{active.experiment_id}', lifecycle='{active.lifecycle_stage}'"
)
```
